### Histogram

#### Histogram แสดงการกระจายตัว(distribution) ของข้อมูลที่เราสนใจ
#### แกน x คือข้อมูลหรือสิ่งที่เราสนใจ โดยจะถูกแบ่งออกเป็นช่วงๆ 
#### แกน Y คือความถี่ในการเกิดข้อมูลนั้นๆหรือสิ่งที่เราสนใจนั้นๆ
##### https://en.wikipedia.org/wiki/Histogram

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/car_mpg.csv'
df=pd.read_csv(url)
df.head()
```

```python
sns.distplot(df['cty']) #draw displot
```

```python
sns.distplot(df['cty'],kde=False) #kde
```

```python
sns.distplot(df['cty'], kde=False)
plt.axvline(x=df.cty.mean()) 
plt.title(f"mean = {df.cty.mean():.2f}, sd = {df.cty.std():.2f}") 
```

```python
sns.distplot(df['cty'], kde=True, label="city")
sns.distplot(df['hwy'], kde=True, label="highway")
plt.legend() 
#draw 2 displot : 1 figure
```

```python
sns.distplot(df['cty'], kde=True, label="city")
plt.show()
sns.distplot(df['hwy'], kde=True, label="highway")#draw 2 displot 
plt.show()
```

```python
fig, ax = plt.subplots(1,2) 
print(ax.shape)
sns.distplot(df['cty'], kde=False, label="city",ax=ax[0])

sns.distplot(df['hwy'], kde=False, label="highway",ax=ax[1])#draw 2 displot 
```

```python
#sharey
fig, ax = plt.subplots(1,2,sharey=True) 
print(ax.shape)
sns.distplot(df['cty'], kde=False, label="city",ax=ax[0])

sns.distplot(df['hwy'], kde=False, label="highway",ax=ax[1])#draw 2 displot 
```

```python
#sharex

fig, ax = plt.subplots(1,2,sharex=True,sharey=True) 
print(ax.shape)
sns.distplot(df['cty'], kde=False, label="city",ax=ax[0])

sns.distplot(df['hwy'], kde=False, label="highway",ax=ax[1])#draw 2 displot 
```
