### Boxplot

#### Boxplot แสดง ค่ากลาง ค่าการกระจาย สัดส่วนข้อมูลที่มากหรือน้อยกว่าค่ากลาง 
#### รวมทั้งข้อมูลที่ผิดปกติ(Outlier) เป็นข้อมูลที่อยู่ห่างจากกลุ่มมากๆ 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/car_mpg.csv'
df=pd.read_csv(url)
df.head()
```

```python
sns.boxplot(data=df[['cty']]) #list
```

```python
sns.boxplot(data=df[['cty','hwy']]) #draw 2 boxplot : 1 figure
```

```python
sns.boxplot(x='cty',y='drv',data=df) #reverse
```

```python
sns.boxplot(y='drv', x='cty',data=df)
plt.ylabel('')
plt.xlabel('city miles per gallon')
plt.title('miles per gallon by driving system')
```

```python
#hue year
sns.boxplot(y='drv', x='cty',data=df,hue='year')
plt.ylabel('')
plt.xlabel('city miles per gallon')
plt.title('miles per gallon by driving system')
```

```python
#figsize
plt.figure(figsize=(10,4))
sns.boxplot(y='drv', x='cty',data=df,hue='year')
plt.ylabel('')
plt.xlabel('city miles per gallon')
plt.title('miles per gallon by driving system')
```

```python
#palette
plt.figure(figsize=(10,4))
sns.boxplot(y='drv', x='cty',data=df,hue='year',palette='Set1')
plt.ylabel('')
plt.xlabel('city miles per gallon')
plt.title('miles per gallon by driving system')
```

```python
#draw 2 plot
sns.catplot(y='drv', x='cty', data=df,
              kind='box',
              col='year') 
```

```python
sns.catplot(y='drv', x='cty', data=df,
              kind='box',
              row='year',
               col='gear')
```
