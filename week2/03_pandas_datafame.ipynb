```python
# https://github.com/mathawanup/learn_dataset/blob/master/mpg.csv
df=pd.read_csv('https://raw.githubusercontent.com/mathawanup/learn_dataset/master/mpg.csv')
df.head()
```

```python
df['cty_kml'] = df['cty'] * .425143707
df.head()
```

```python
df['cty_kml'] = df['cty_kml'].round(1)
df.head()
```

```python
df['trans_am'] = df.trans.str[0]
df.head()
```

```python
filepath = 'F:\61020593\data_mpg.csv' # $ conda install openpyxl
df.to_csv(filepath)
```

```python
filepath = 'F:\61020593\data_mpg.xlsx' #  $ conda install openpyxl  ติดตั้ง Anaconda prompt
writer=pd.ExcelWriter(filepath)
df.to_excel(writer)
writer.save()
```
