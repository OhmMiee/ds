## lineplot 

**a line plot with the possibility of several semantic groupings. The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/tuition_graduate.csv'
df=pd.read_csv(url)
# df.head()
df
```

```python
# unique ดูว่ามีกี่คณะ  
df.school.unique() 
```

```python
#ดูว่าแต่ละปีมีค่าใช้จ่ายเท่าไหร่ แต่ละคณะ(school)
sns.lineplot(data=df,
             x='academic.year', y='cost', 
             hue='school')  
```

```python
#จัดการ label
plt.figure(figsize=(8,6))
sns.lineplot(data=df,
             x='academic.year', y='cost', 
             hue='school')  
```

```python
#อยากทราบคณะที่ต้องการ
df[df.school.str.match('Business|Design|Law')] 
```

```python
#แสดงกราฟคณะที่ต้องการ
plt.figure(figsize=(8,6))
sns.lineplot(data=df[df.school.str.match('Business|Design|Law')],
             x='academic.year', y='cost', 
             hue='school')  
```

```python
#แสดงข้อมูลหนึ่งคณะต่อหนึ่งกราฟ
sns.relplot(data=df,
             x='academic.year', y='cost', 
             hue='school',
             col='school',
             kind='line',)
```

```python
#แสดงข้อมูลหนึ่งคณะต่อหนึ่งกราฟ และกำหนด col กราฟที่ต้องการแสดง
sns.relplot(data=df,
             x='academic.year', y='cost', 
             hue='school',
             col='school',
             kind='line',
             col_wrap=3,
           )
```

```python
#set font size
sns.set_context('notebook',font_scale=1.6)
sns.relplot(data=df,
             x='academic.year', y='cost', 
             hue='school',
             col='school',
             kind='line',
             col_wrap=3,
           )
```
