### scatter plot 

#### แผนภาพการกระจาย (Scatter Plot) แสดงความสัมพันธ์ของข้อมูล 2 ตัว  โดยค่าของตัวแปรทัังสองมาเขียนกราฟ
#### แต่ละจุดในแผนภาพการกระจายคือ ค่า x และ y เป็นคู่ๆ 

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/car_mpg.csv'
df=pd.read_csv(url)
df.head()
```

```python
sns.scatterplot(x="displ", y="cty", data=df)
```

```python
sns.scatterplot(x="displ", y="cty", data=df, hue='drv')
```

```python
sns.relplot(x="displ", y="cty", data=df) 
```

```python
sns.relplot(x="displ", y="cty", hue="drv", data=df)
```

```python
df['class'].unique()
```

```python
g = sns.relplot(x="displ", y="cty", hue="drv", col="class", 
                data=df)
```

```python
g = sns.relplot(x="displ", y="cty", hue="drv", col="class", col_wrap=3,
                data=df)
```

```python
df[df['class'].str.match('compact|midsize', case=False)]
```

```python
sns.relplot(x="displ", y="cty", hue="drv", col="class", 
                data=df[df['class'].str.match('compact|midsize', case=False)]
           )
```

```python
df.head()
```

```python
g = sns.relplot(x="displ", y="cty", hue="drv", 
                col="class", row='gear',
                data=df[df['class'].str.match('compact|midsize', case=False)])
```
