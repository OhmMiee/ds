## lineplot 

**a line plot with the possibility of several semantic groupings. The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/tuition_graduate.csv'
df=pd.read_csv(url)
df
```

```python
df.school.unique() 
```

```python
sns.lineplot(data=df,
             x='academic.year', y='cost', 
             hue='school')  
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=df, x='academic.year', y='cost', hue='school')  
```

```python
data1 = df[df.school.str.match('Business|Design|Law')] 
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=data1 , x='academic.year', y='cost', hue='school')  
```

```python
# plt.figure(figsize=(10,10))
sns.relplot(data=df, x='academic.year', y='cost', hue='school', col='school', kind='line')  
```

```python
sns.relplot(data=df, x='academic.year', y='cost', hue='school', col='school', kind='line', col_wrap=3)  
```

```python
sns.set_context('notebook', font_scale=1.6)
sns.relplot(data=df, x='academic.year', y='cost', hue='school', col='school', kind='line', col_wrap=3)  
```
