#### ให้นิสิตจัดการกับ Missing values ต่อไปนี้

```python
import pandas as pd
```

```python
# https://github.com/mathawanup/learn_dataset/blob/master/property_data.csv
df=pd.read_csv('https://github.com/mathawanup/learn_dataset/raw/master/property_data.csv')
df
```

```python
df.describe()
```

**1.แทนค่า Missing value ในคอคัมภ์ ST_NUM และ SQ_FT ด้วยค่าเฉลี่ย**

```python
df.ST_NUM.fillna(df.ST_NUM.mean(), inplace=True)
df
```

```python
df.SQ_FT.fillna(df.SQ_FT.mean(), inplace=True)
df
```

**2. แทนค่า Missing value ในคอคัมภ์ NUM_BATH ด้วย 2**

```python
df.NUM_BATH.fillna(2, inplace=True)
df
```

**3.แทนที่ Missing value ในคอลัมภ์ OWN_OCCUPIED**

```python
df.OWN_OCCUPIED.value_counts()
```

```python
df.OWN_OCCUPIED.fillna('Y', inplace=True)
df
```
