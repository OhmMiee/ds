### ให้นิสิตเปลี่ยนข้อมูลจาก wide format ให้เป็น long format 

```python
import pandas as pd
```

## Wide format

**อ่านข้อมูล จากลิงค์ต่อไปนี้ https://github.com/mathawanup/learn_dataset/blob/master/subjgroup.csv**

```python
df=pd.read_csv('https://raw.githubusercontent.com/mathawanup/learn_dataset/master/subjgroup.csv')
df
```

**1.เปลี่ยนข้อมูลให้เป็น long format**

**หมายเหตุ melt()  คอลัมภ์ Application	,Programming	,Mobile**

```python
df2=pd.melt(df, id_vars=['Subject','Group'])
df2
```

**2.สร้าง pivot table  แบบ Multple Index**

```python
df2.columns
```

```python
df3=df2.pivot_table(index=['Subject','Group'], columns='variable', values='value',margins=True, aggfunc='sum')
df3
```
