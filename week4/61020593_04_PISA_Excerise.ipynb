**กำหนดให้คอลัมภ์ประเทศ (Country) เป็น index ของ dataframe**

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/pisa_score.csv'
df = pd.read_csv(url, index_col = 'Country')
df
```

**หาคะแนนเฉลี่ยของแต่ละวิชา**

```python
df.mean(axis=0)
```

```python
df.Maths.mean()
```

```python
df.Science.mean()
```

```python
df.Reading.mean()
```

**แสดงประเทศที่มีคะแนน Maths น้อยกว่าค่าเฉลี่ย**

```python
df.loc[(df.Maths<=df.Maths.mean())]
```

**แสดงประเทศที่มีคะแนน Science มากกว่าหรือเท่ากับค่าเฉลี่ย**

```python
df.loc[(df.Science>=df.Science.mean())]
```
